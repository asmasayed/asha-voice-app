```python
#import all the libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
#plot function: (points along x axis, along y axis) it will draw a simple line graph
x= [0,2,4,5,6]
y=[1,2.7,5,4,6]

#resize the figure
plt.figure(figsize=(4,3),dpi=150)

#modify the graph; add label to the x,y axis using label='value' etc
#plt.plot(x,y,label='2x',color='red',linewidth=2.4,marker='.',markersize=10,markeredgecolor='black',linestyle='dashed')

#shortcut way to modify this is using fmt=[color][marker][line]
plt.plot(x,y,'bs:',label='2x')
plt.title('First Graph', fontdict={'fontsize':20,'fontname':'Sans-serif'})

plt.xlabel('X Axis')
plt.ylabel('Y Axis')

#change the values on x,y axis:
#plt.xticks([0,1,2,3,4,5,6])
#plt.yticks([0,1,2,3,4,5,6])

#Line number 2
x2=np.arange(0,4.5,0.5)
plt.plot(x2[:4],x2[:4]**2,'r',label='x^2')
plt.plot(x2[3:],x2[3:]**2,'r:',label='x^2')

# This looks for the 'label' on each plot and displays them in a LEGEND box
plt.legend()

#to save a figure, bbox inches makes sure matplotlib saves the grapg without cutting any labels or ticks
#plt.savefig('myfirstgraph.png',dpi=200,bbox_inches="tight", pad_inches=0.2)

plt.show()
```

### Bar Chart

```python
#resize the figure
plt.figure(figsize=(3,2),dpi=150)

values=['a','b','c','d']
labels=[200,300,240,300]

#Plot the bar chart
bars= plt.bar(values,labels)

#plot patterns on bars
patterns=['/','O','*','+']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

#Or simply set each bar individually
#bars[0].set_hatch('/')
#bars[1].set_hatch('O')
#bars[2].set_hatch('*')
#bars[3].set_hatch('+')
plt.show()
```

### Real life dataset Examples


### Line Graph

```python
#Load the data
gas=pd.read_csv('gas_prices.csv')
gas.head()
```

```python
plt.figure(figsize=(5,4),dpi=150)
plt.title('Gas Prices In USD Over Year',fontdict={'fontweight':'bold'})
for country in gas:
    if country != 'Year':
        plt.plot(gas.Year,gas[country],'.-',label=country)
plt.xticks(gas.Year[::3])
#if you want to see a specific year
#gas->dataframe gas.Year->Series [::3]->Every third value .tolist->converts the series to list so you can add another list +->appends the [2011] at end
plt.xticks(gas.Year[::3].tolist()+[2011])

#move legend outside the box
plt.legend(bbox_to_anchor=(1.05,1),loc='upper left')
plt.show()
```

### Histograms

```python
#Load the file
fifa=pd.read_csv('fifa_data.csv')
fifa.head()
```

```python
plt.title('Distribution of player skills in FIFA')
bins=[40,50,60,70,80,90,100]
plt.hist(fifa.Overall,bins=bins,color="#0cb000")
plt.xticks(bins)
plt.xlabel('Skill level')
plt.ylabel('No of players')


plt.show()
```

### Pie Chart

```python
#.pie-> Always accept array-like numeric values or panda series
#s=pd.Series([10,20,30])
#plt.pie(s)

#Pie Chart 1
#Foot Preference of Fifa Players
#left=fifa.loc[fifa['Preferred Foot']=='Left']
#left=len(left)
#right=fifa.loc[fifa['Preferred Foot']=='Right']
#right=len(right)

#plt.pie([left,right],labels=['Left','Right'],colors=['red','orange'],autopct='%.2f %%')
#plt.title("Foot Preference of Fifa Players")
#plot.show()

#Pie Chart 1
#Weight of Fifa Players
fifa.columns
# Since weight is a string extract only the number
fifa['Weight']=[int(x.replace('lbs','')) if type(x)==str else x for x in fifa.Weight]
light        = fifa.loc[(fifa['Weight'] > 120) & (fifa['Weight'] < 150)]
light_medium = fifa.loc[(fifa['Weight'] >= 150) & (fifa['Weight'] < 160)]
medium       = fifa.loc[(fifa['Weight'] >= 160) & (fifa['Weight'] < 185)]
medium_heavy = fifa.loc[(fifa['Weight'] >= 185) & (fifa['Weight'] < 200)]
heavy        = fifa.loc[fifa['Weight'] >= 200]

w=[len(light),len(light_medium),len(medium),len(medium_heavy),len(heavy)]
labels = ['Light', 'Light-Medium', 'Medium', 'Medium-Heavy', 'Heavy']
explode=[.1,0,0,.1,.1]
plt.title('Weight distribution of Players')
plt.pie(w,labels=labels,autopct='%.2f %%',pctdistance=0.8,explode=explode)
plt.show()
```

### Box and wiskers charts

```python
#clubs and the relative players score
plt.figure(figsize=(4,6))
fifa.head()
barcelona=fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid=fifa.loc[fifa.Club == 'Real Madrid']['Overall']
plt.xlabel('Clubs')
plt.ylabel('Player Ratings')
plt.title('Player ratings in clubs')
label=['Barcelona','Real Madrid']
boxes=plt.boxplot([barcelona,madrid],tick_labels=label,patch_artist=True,medianprops={'linewidth':3})
for box in boxes['boxes']:
    box.set(color='#43abd4',linewidth=2,facecolor='#e0e0e0')
plt.show()
```
